```python
from sentence_transformers import SentenceTransformer, models
import gradio as gr
import numpy as np
```

```python
model = SentenceTransformer('chienpham/sts-multilingual-mpnet-base-v2')
```

```python
def demo(sentence_1, sentence_2):
  v1 = model.encode(sentence_1)
  v2 = model.encode(sentence_2)
  p = np.dot(v1,v2)/(np.linalg.norm(v1)*np.linalg.norm(v2))*5
  return "Yes" if p > 3.75 else "No"
```

```python
iface = gr.Interface(fn=demo, inputs=["text", "text"], outputs="text")
iface.launch()
```
